```python
df_top_10 = df.X1.value_counts().sort_values(ascending = False).head(10).index
```

```python
df_top_10 = list(df_top_10)
df_top_10
```

```python
for i in df_top_10:
    df[i] = np.where(df['X1'] == i,1,0)
```
